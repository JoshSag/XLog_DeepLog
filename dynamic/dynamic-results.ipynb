```python
import nbformat as nbf
import pandas as pd
import re
```

```python
dfs = list()
def get_params(nb):
    d=nbf.read(nb, nbf.NO_CONVERT)
    for c in d.cells:
        if "### Parameters" in c.source:
            vocabulary_size = re.search("vocabulary_size = (\d+)", c.source).groups()[0]
            num_patterns = re.search("num_patterns = (\d+)", c.source).groups()[0]
            num_new_patterns = re.search("num_new_patterns = (\d+)", c.source).groups()[0]
            return int(vocabulary_size), int(num_patterns), int(num_new_patterns)
            
def get_result_from_nb(nb):
    try:
        global dfs
        d=nbf.read(nb, nbf.NO_CONVERT)
        cur_df = pd.read_html(d.cells[-1].outputs[-1]["data"]["text/html"])[0]
        cur_df["file"] = nb

        vocabulary_size, num_patterns, num_new_patterns = get_params(nb)
        cur_df["vocabulary_size"] = vocabulary_size
        cur_df["num_patterns"] = num_patterns
        cur_df["num_new_patterns"] = num_new_patterns
        dfs.append(cur_df)
    except:
        pass
```

```python
nbs = ["dynamic-1.ipynb", "dynamic-2.ipynb", "dynamic-3.ipynb", "dynamic-4.ipynb", "dynamic-5.ipynb", "dynamic-6.ipynb"]
for nb in nbs:
    get_result_from_nb(nb)
df = pd.concat(dfs)
df.drop("Unnamed: 0",axis=1,inplace=True)
```
