```python
import nbformat as nbf
import pandas as pd
import re
```

```python
dfs = list()
def get_params(nb):
    d=nbf.read(nb, nbf.NO_CONVERT)
    for c in d.cells:
        if "### Parameters" in c.source:
            vocabulary_size = re.search("vocabulary_size = (\d+)", c.source).groups()[0]
            num_patterns = re.search("num_patterns = (\d+)", c.source).groups()[0]
            return int(vocabulary_size), int(num_patterns)
            
def get_result_from_nb(nb):
    global dfs
    d=nbf.read(nb, nbf.NO_CONVERT)
    cur_df = pd.read_html(d.cells[-1].outputs[-1]["data"]["text/html"])[0]
    cur_df["file"] = nb
    
    vocabulary_size, num_patterns = get_params(nb)
    cur_df["vocabulary_size"] = vocabulary_size
    cur_df["num_patterns"] = num_patterns
    dfs.append(cur_df)
```

```python
nbs = ["static-1.ipynb", "static-2.ipynb", "static-3.ipynb", "static-4.ipynb", "static-5.ipynb", "static-6.ipynb", "static-7.ipynb"]
for nb in nbs:
    get_result_from_nb(nb)
df = pd.concat(dfs,ignore_index=True)
df.rename(columns = {"Unnamed: 0": "g"}, inplace=True)
```
